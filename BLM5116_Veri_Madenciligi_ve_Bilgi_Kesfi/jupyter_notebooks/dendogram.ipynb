```python
dots = {
    'p': [1, 3],
    'q': [3, -2],
    'r': [2, 4],
    's': [5, 2],
    't': [-6, 1],
    'u': [7, 3],
}

# Path: dendogram.ipynb

def distance(a, b):
    return round(((a[0] - b[0])**2 + (a[1] - b[1])**2) ** 0.5, 1)

distance_matrix = {
    'p': {
        'p': 0,
        'q': distance(dots['p'], dots['q']),
        'r': distance(dots['p'], dots['r']),
        's': distance(dots['p'], dots['s']),
        't': distance(dots['p'], dots['t']),
        'u': distance(dots['p'], dots['u']),
    },
    'q': {
        'p': distance(dots['q'], dots['p']),
        'q': 0,
        'r': distance(dots['q'], dots['r']),
        's': distance(dots['q'], dots['s']),
        't': distance(dots['q'], dots['t']),
        'u': distance(dots['q'], dots['u']),
    },
    'r': {
        'p': distance(dots['r'], dots['p']),
        'q': distance(dots['r'], dots['q']),
        'r': 0,
        's': distance(dots['r'], dots['s']),
        't': distance(dots['r'], dots['t']),
        'u': distance(dots['r'], dots['u']),
    },
    's': {
        'p': distance(dots['s'], dots['p']),
        'q': distance(dots['s'], dots['q']),
        'r': distance(dots['s'], dots['r']),
        's': 0,
        't': distance(dots['s'], dots['t']),
        'u': distance(dots['s'], dots['u']),
    },
    't': {
        'p': distance(dots['t'], dots['p']),
        'q': distance(dots['t'], dots['q']),
        'r': distance(dots['t'], dots['r']),
        's': distance(dots['t'], dots['s']),
        't': 0,
        'u': distance(dots['t'], dots['u']),
    },
    'u': {
        'p': distance(dots['u'], dots['p']),
        'q': distance(dots['u'], dots['q']),
        'r': distance(dots['u'], dots['r']),
        's': distance(dots['u'], dots['s']),
        't': distance(dots['u'], dots['t']),
        'u': 0,
    },
}

print(distance_matrix)
```

```python
import pandas as pd

df = pd.DataFrame(distance_matrix)
df
```

```python
dots_df = pd.DataFrame(dots).T
dots_df.columns = ['x', 'y']
dots_df
```

```python
# draw dendogram
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

linked = linkage(dots_df, 'single')

labelList = list(dots_df.index)

plt.figure(figsize=(20, 7))
dendrogram(linked,
            orientation='right',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
linked = linkage(dots_df, 'complete')

labelList = list(dots_df.index)

plt.figure(figsize=(20, 7))
dendrogram(linked,
            orientation='right',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```
