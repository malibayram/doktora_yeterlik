```python
# import plotting library
import matplotlib.pyplot as plt
dots = [[0, 0, 1], [0.3, -0.1, 1], [-0.12, 0.2, 1], [-0.3, -0.1, 1],[0,0.3, 1], [0.3, 0.2, 1], [-0.12, -0.1, 1], [-0.3, 0.2, 1],[1, 1, -1], [-1, -1, -1], [0, 0.7, -1], [0.6, 0.8, -1], [1, -1, -1], [-1, 1, -1], [0, -0.7, -1], [-0.6, -0.8, -1]]

# plot the dots
for dot in dots:
    if dot[2] == 1:
        plt.plot(dot[0], dot[1], 'ro')
    else:
        plt.plot(dot[0], dot[1], 'bo')

# show the plot
plt.show()
```

```python
class CirclePerceptron:
    def __init__(self, w1=0.0, b=0.0):
        self.w1 = w1 # as radius of circle
        self.b = b # as bias

    def predict(self, x): # x is a list of two numbers representing a point
        # check if the point is inside the circle
        # also consider the bias
        return 1 if (x[0] ** 2 + x[1] ** 2 - self.w1 ** 2 + self.b) >= 0 else -1

    # train data and update the weight to make the circle smaller or bigger
    def train(self, x_dots, epochs=10, learning_rate=0.01):
        for epoch in range(epochs):
            # update weight and bias for each epoch
            print("Epoch: ", epoch)
            #draw the circle
            self.plot(x_dots)
            for x in x_dots:
                y = self.predict(x)
                if y == x[2]:
                    pass
                elif y == 1:
                    self.w1 -= learning_rate                    
                else:
                    self.w1 += learning_rate
                # burada ayrca emberin iindeki noklarn ortas hesaplanp
                # emberin merkezi bu noktaya ekilerek emberin kaymas salanr.
                
    def plot(self, x_dots,):
        # plot the x dots red and y dots blue
        for x in x_dots:
            if x[2] == 1:
                plt.plot(x[0], x[1], 'ro')
            else:
                plt.plot(x[0], x[1], 'bo')
        # plot the circle
        circle = plt.Circle((0, 0), self.w1, color='g', fill=False)
        plt.gcf().gca().add_artist(circle)
        plt.show()
circle_perceptron = CirclePerceptron()
circle_perceptron.train(dots)
```
